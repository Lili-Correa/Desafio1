```
#imports
import altair as alt
import numpy as np
import pandas as pd
#definir os dados
x  = [1, 2, 3, 4, 5, 6, 7, 8, 9]
y1 = [2, 1, 0, 3, 0, 5, 1, 3, 1]
y2 = [3, 5, 7, 5, 3, 5, 7, 4, 2]

#Criar df
source = pd.DataFrame({ 'x': x,'y1': y1, 'y2': y2 })
#criar uma base
base = alt.Chart(source).encode(x='x')
#criar uma camada com 2 linhas
alt.layer(
    base.mark_line(color='black').encode(y='y1'),
    base.mark_line(color='green').encode(y='y2')
).interactive()
```

##PLOTLY

```
# imports
import plotly.express as px
import plotly.graph_objs as go
import pandas as pd 
import numpy as np

#definir os dados
x  = [1, 2, 3, 4, 5, 6, 7, 8, 9]
y1 = [2, 1, 0, 3, 0, 5, 1, 3, 1]
y2 = [3, 5, 7, 5, 3, 5, 7, 4, 2]

#criar um dataframe
df = pd.DataFrame({'x': x,'y1': y1,'y2': y2})

#criar uma figura
fig = go.Figure()
#criar a primeira linha
fig.add_trace(go.Scatter(x=df['x'], y=df['y1'].values,
                             name = "y1",
                             mode = 'markers+lines',
                             line=dict(shape='linear',color='rgb(0,0,0)', width=6),
                             connectgaps=True) )
# criar a segunda linha
fig.add_trace(go.Scatter(x=df['x'], y=df['y2'].values,
                             name = "y2",
                             mode = 'markers+lines',
                             line=dict(shape='linear',color='rgb(0,255,0)', width=6),
                             connectgaps=True) )
#criar o t√≠tulo e r√≥tulos dos eixos
fig.update_layout(
        title={
        'text': "PLOTLY",
        'y':0.9,
        'x':0.5,
        'xanchor': 'center',
        'yanchor': 'top'},
            xaxis_title="X Axis Title",
    yaxis_title="Y Axis Title")
fig.show()
```

##BOKEH

```
import numpy as np
from bokeh.plotting import figure, show
from bokeh.io import output_notebook

# Mostra o gr√°fico dentro do Notebook.
output_notebook()

# definir os dados
x  = [1, 2, 3, 4, 5, 6, 7, 8, 9]
y1 = [2, 1, 0, 3, 0, 5, 1, 3, 1]
y2 = [3, 5, 7, 5, 3, 5, 7, 4, 2]

#criar una figura
p = figure(title="BOKEH", x_axis_label="Eixo x", y_axis_label="Eixo y")

# graficar linha de y1
p.line(x, y1, legend_label="y1.", line_width=2, line_color='black')
# graficar linha de y2
p.line(x, y2, legend_label="y2", line_width=2, line_color='green')

# show the results
show(p)
```
